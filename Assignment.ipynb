# Q2

Write a program that accepts a sentence and calculate the number of letters and digits. Suppose the following input is supplied to the program:
hello world! 123
Then, the output should be:
LETTERS 10 DIGITS 3

```python
import re

#line = 'hello world! 123'

line = input('Enter a Line: \n')

letters = 0
digits = 0

l = re.compile('[a-zA-Z]')
d = re.compile('[0-9]')

for word in line.split():
    for ltr in word:
        #count letter if it's a letter via regex matching
        if l.match(ltr):
            letters+=1
        
        #elif count number if it's number via regex matching
        elif d.match(ltr):
            digits+=1
            
print('LETTERS: {} \nDIGITS: {}'.format(letters,digits))

## Target Output
# LETTERS 10
# DIGITS 3

```

# Q3
Build a function that takes six passwords and checks which one is correct and returns all that match and don't and why.

```python
import re

# Declare Regulr expressions
password_regex = re.compile('^(?=.*[a-z])(?=.*[A-Z])(?=.*[0-9])(?=.*[$#@]).{6,12}$')
pw_regex_smallatoz = re.compile('(?=.*[a-z])')
pw_regex_bigatoz =  re.compile('(?=.*[A-Z])')
pw_regex_digits =  re.compile('(?=.*[0-9])')
pw_regex_special =  re.compile('(?=.*[$#@])')
 
# Use of Regular Expressions with Positive look ahead to check if:
# (?=.*[a-z])    At least one lower case letter exists
# (?=.*[A-Z])    At least one upper case letter exists
# (?=.*[0-9])    At least one digit exists
# (?=.*[$#@])    At least one special character exists
# .{6,12}        And finally if its minimum 6 char and max 12 char

# This is the main function that checks the conditions individually
def check_failure_reason(failed_password):
    failure_reason = []
 
    failure_reason = check_smallatoz(failed_password, failure_reason)
    failure_reason = check_bigatoz(failed_password, failure_reason)
    failure_reason = check_digits(failed_password, failure_reason)
    failure_reason = check_special_char(failed_password, failure_reason)
    failure_reason = check_length(failed_password, failure_reason)
   
    return failure_reason
 
# Check whether the password passes or fails small a to z check.
# Must have atleast one lower case letter.
# Then update the [failure] reason if it fails.
def check_smallatoz(password, reason):
 
    if pw_regex_smallatoz.match(password):
        print('Password Met small a-z check')
 
    else:
        reason.append("Password Failed small a-z check")
 
    return reason
 
# Check whether the password passes or fails big A to Z check.
# Must have atleast one upper case letter.
# Then update the [failure] reason if it fails.
def check_bigatoz(password, reason):
 
    if pw_regex_bigatoz.match(password):
        print('Password Met big A-Z check')
 
    else:
        reason.append("Password Failed big A-Z check")
 
    return reason

# Check whether the password passes or fails the digits check.
# Must be atleast 1 digit.
# Then update the [failure] reason if it fails.
def check_digits(password, reason):
 
    if pw_regex_digits.match(password):
        print('Password Met digits check')
 
    else:
        reason.append("Password Failed digits check")
 
    return reason

# Check whether the password passes or fails the digits check.
# Must have atleast 1 special character.
# Then update the [failure] reason if it fails.
def check_special_char(password, reason):
 
    if pw_regex_special.match(password):
        print('Password Met special character check')
 
    else:
        reason.append("Password Failed special character check")
 
    return reason

# Check whether the password passes or fails the digits check.
# Must be atleast 6 charcters and no more than 12 characters long.
# Then update the [failure] reason if it fails.
def check_length(password, reason):
 
    pw_length = len(password)
 
    if (pw_length >=6 and pw_length <=12):
        print('Password Met Length check')
 
    else:
        reason.append("Password Failed Length check, ie min 6 max 12 characters")
 
    return reason
 
# User Entry
line = input('Enter some passwords seperated by a comma \',\': \n')

# Declare list variables for use later (is this necessary?)
pass_list = []
fail_list = []
 
#line = 'ABd1234@1,a F1#,2w3E*,2We3345, ABd1234@2, ABD1234@9'
line_list = line.split(',')
 
# For each word check if
for i in line_list:
    if password_regex.match(i.strip()):
        pass_list.append(i.strip())
    else:
        fail_list.append(i.strip())
 
print("\nPassed:")
print(*pass_list, sep = ", ",end='\n\n')

# Check the failure list - won't need to do this if there aren't any failures
for f in fail_list:
    print("\nFailed: ", f)
    fail_reason = check_failure_reason(f)
    print(*fail_reason,sep = ", ",end='\n')
```
